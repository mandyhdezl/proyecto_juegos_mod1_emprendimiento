```python
#Si el usuario ingresaba algo incorrecto el programa solo mostraba "Respuesta no vÃ¡lida" y seguÃ­a con la siguiente ronda. 
#Ahora, se usa un while True para asegurarnos de que el usuario solo pueda elegir "piedra", "papel" o "tijera". 
#Valido la elecciÃ³n del usuario con una funciÃ³n para evitar errores.

import random                                                                                 
def obtener_eleccion():
    opciones = ['piedra', 'papel', 'tijera']                                     #Antes el juego con for limitaba el juego a 3 rondas. 
                                                                                  #Ahora, con while, el juego sigue hasta que alguien llegue al objetivo.

    while True:                                                                 
        eleccion = input('Elige piedra, papel o tijera: ').lower()
        if eleccion in opciones:
            return eleccion
        print('OpciÃ³n no vÃ¡lida. IntÃ©ntalo de nuevo.')


usuario_puntos = 0                                                                  #Defino puntos objetivo para ganar. 
puntos_objetivo = 3
pc_puntos = 0

print('Â¡Â¡Â¡Bienvenido al juego de Piedra, Papel y Tijera!!!')
print(f'El primero en llegar a {puntos_objetivo} puntos gana.\n')

while usuario_puntos < puntos_objetivo and pc_puntos < puntos_objetivo:
    usuario = obtener_eleccion()                                                   #Antes pedÃ­amosla entrada del usuario directamente con input(), 
                                                                                    #ahora usamos la funciÃ³n para asegurarnos de que sea vÃ¡lida.
    pc = random.choice(['piedra', 'papel', 'tijera'])

    print(f'TÃº eliges: {usuario}')
    print(f'Tu contrincante eligiÃ³: {pc}')
  
    if usuario == pc:
        print('Has empatado ğŸ¤“')                                                  
    elif (usuario == 'piedra' and pc == 'tijera') or\
         (usuario == 'tijera' and pc == 'papel') or\
         (usuario == 'papel' and pc == 'piedra'):
        print('Â¡Has ganado esta ronda! ğŸ‰')
        usuario_puntos += 1                                                        #Se muestran los puntajes de cada ronda, para que el usuario sepa.
    else:
        print('Has perdido esta ronda. ğŸ˜')
        pc_puntos += 1

    print(f'Puntaje: TÃº {usuario_puntos} - {pc_puntos} Computadora\n')

#Verifico el ganador final.                                                        #Anunciamos al ganador final de forma clara.
if usuario_puntos > pc_puntos:
    print('Â¡Felicidades! Has ganado el juego ğŸ†')
else:
    print('Tu oponente ha ganado esta vez. Â¡Sigue practicando!Â ğŸ¤–')
```
